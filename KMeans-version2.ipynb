```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
#載入資料
df = pd.read_csv('finish_data2.csv')
df.head()
```

```python
#將不必要的欄位刪除
df.drop(columns=['product_name', 'product_tags','INTEGER', 'region', 'product_price'], inplace=True)
```

```python
#確認缺失值
df.info()
```

此data無缺失值

```python
#將country欄位二分為台灣(1)及非台灣(0)
df['Country'] = df['country'].apply(lambda x: 1 if x == 'Taiwan' else 0)
df.head(2)
```

```python
#將country刪除 
df = df.drop(['country'], axis=1)
```

```python
#查詢source欄位裡的各類別數量
df['source'].value_counts()
```

```python
#將從類似拜訪網站的管道合在一起
df['source'] = df['source'].str.replace('l.instagram.com', 'instagram.com')

df['source'] = df['source'].str.replace('m.facebook.com', 'facebook.com')
df['source'] = df['source'].str.replace('l.facebook.com', 'facebook.com')
df['source'] = df['source'].str.replace('lm.facebook.com', 'facebook.com')
df['source'] = df['source'].str.replace('lfacebook.com', 'facebook.com')

df['source'] = df['source'].str.replace('google', 'other')
df['source'] = df['source'].str.replace('reurl.cc', 'other')
df['source'] = df['source'].str.replace('youtube.com', 'other')
df['source'] = df['source'].str.replace('m.youtube.com', 'other')
df['source'] = df['source'].str.replace('haloepoch.easy.co', 'other')
df['source'] = df['source'].str.replace('admin.easystore.co', 'other')
df['source'] = df['source'].str.replace('m.other', 'other')
```

-只保留數量前三多的管道，分別為instagram.com、(direct)、facebook.com，其餘管道被分為other

```python
df['source'].value_counts()
```

```python
#因sub_categ裡的類別太多，故做整理
df['sub_categ'] = df['sub_categ'].str.replace('貝蕾帽', '其餘帽子')
df['sub_categ'] = df['sub_categ'].str.replace('報童帽', '其餘帽子')
df['sub_categ'] = df['sub_categ'].str.replace('狩獵帽', '其餘帽子')


df['sub_categ'] = df['sub_categ'].str.replace('包包', '配件')
df['sub_categ'] = df['sub_categ'].str.replace('項鍊', '配件')
df['sub_categ'] = df['sub_categ'].str.replace('襪子', '配件')
df['sub_categ'] = df['sub_categ'].str.replace('圍巾', '配件')
df['sub_categ'] = df['sub_categ'].str.replace('戒指', '配件')

df['sub_categ'] = df['sub_categ'].str.replace('衣服', '上身')
df['sub_categ'] = df['sub_categ'].str.replace('外套', '上身')

df['sub_categ'] = df['sub_categ'].str.replace('裙子', '下身')
df['sub_categ'] = df['sub_categ'].str.replace('褲子', '下身')
```

-帽子只保留前三多瀏覽數量的類別，分別為老帽&棒球帽、鐘形帽&漁夫帽、毛帽，其餘類別被分為其餘帽子

-包包、項鍊、襪子、圍巾、戒指被分為配件

-衣服、外套被分為上身

-裙子、褲子被分為下身

```python
#查詢sub_categ欄位裡的各類別數量
df['sub_categ'].value_counts()
```

```python
#將類別型資料做獨熱編碼 
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse=False)
encoded_columns = encoder.fit_transform(df[['first_visit', 'source', 'sub_categ', 'price_category']])

encoded_column_names = encoder.get_feature_names_out(['first_visit', 'source', 'sub_categ', 'price_category'])

encoded_df = pd.DataFrame(encoded_columns, columns=encoded_column_names)
```

```python
df2 = pd.concat([df.drop(columns=['first_visit', 'source', 'sub_categ', 'price_category']), encoded_df], axis=1)
df2
```

```python
df2.describe()
```

# KMeans

```python
#因user_pseudo_id為識別欄位，會影響KMeans結果，故將其刪除
df2 = df2.drop(['user_pseudo_id'], axis=1)
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(init='k-means++', n_clusters=6, random_state=0)
kmeans.fit(df2)
```

```python
labels = kmeans.labels_
df2['cluster'] = labels
```

```python
df2
```

```python
for i in range(6):  # n_clusters的數量
    print(f"\nCluster {i} statistics:")
    print(df2[df2['cluster'] == i].describe())
```
